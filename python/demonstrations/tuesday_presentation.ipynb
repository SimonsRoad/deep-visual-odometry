```python
import json
import numpy as np
import matplotlib.pyplot as plt

def plot_losses(net, filename='history_0.json', model_dir='/home/sexy/source/deep-visual-odometry/models/'):
    #Load all data from json
    with open(model_dir+net+"/"+filename) as data_file:    
        data = json.load(data_file)

    #Set up data for graphing
    losses = [data["loss"],data["val_loss"]]
    num_epoch = [x+1 for x in range(len(losses[0]))]

    #Plot training loss per epoch versus validation loss per epoch
    plt.plot(num_epoch, losses[0], label='Training')
    plt.plot(num_epoch, losses[1], label='Validation')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

def plot_comparison(filename1, filename2, filename3, model_dir='/home/sexy/source/deep-visual-odometry/models/'):
    #Load all data from json
    with open(model_dir+filename1) as data_file:    
        data1 = json.load(data_file)
    with open(model_dir+filename2) as data_file:    
        data2 = json.load(data_file)
    with open(model_dir+filename3) as data_file:    
        data3 = json.load(data_file)
    
    #Set up data for graphing
    losses1 = [data1["loss"],data1["val_loss"]]
    losses2 = [data2["loss"],data2["val_loss"]]
    losses3 = [data3["loss"],data3["val_loss"]]
    num_epoch = [x+1 for x in range(len(losses1[0]))]

    #Plot training loss per epoch versus validation loss per epoch
    f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
    
    ax1.plot(num_epoch, losses1[0], label='Training')
    ax1.plot(num_epoch, losses1[1], label='Validation')
    ax1.set_title('Global')
    
    ax2.plot(num_epoch, losses2[0], label='Training')
    ax2.plot(num_epoch, losses2[1], label='Validation')
    ax2.set_title('Relative')
    
    ax3.plot(num_epoch, losses3[0], label='Training')
    ax3.plot(num_epoch, losses3[1], label='Validation')
    ax3.set_title('Quaternion')
    
    plt.show()
```
