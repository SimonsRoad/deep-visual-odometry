```python
# Y axis is elevation
from os import listdir
import numpy as np
import pykitti
import matplotlib.pyplot as plt
from scipy.misc import imread

#Loads x,y,z translation and Euler x,y,z into a Python list
def load_trans_rot_matrix():
    
    directoryL = "/home/sexy/Documents/dataset/"
    list_Y = []
    
    for i in range(11): #One for each sequence
        data = pykitti.odometry(directoryL, str(i).zfill(2))
        data.load_poses()

        print (data.T_w_cam0[:2])
        Y = np.zeros((len(data.T_w_cam0),6))
        Y_delta = np.zeros((len(data.T_w_cam0),6))
        Y_d = np.zeros((len(data.T_w_cam0),7))
        
        #Array with original total distances and total relative x,y,z angles calculated from rotational matrix
        for j in range(len(data.T_w_cam0)):
            Y[j,:3] = data.T_w_cam0[j][:-1,-1]
            Y[j,3] = np.arctan2(data.T_w_cam0[j][2,1], data.T_w_cam0[j][2,2])
            Y[j,4] = np.arctan2(-data.T_w_cam0[j][2,0], np.sqrt(np.square(data.T_w_cam0[j][2,1]) + np.square(data.T_w_cam0[j][2,2])))
            Y[j,5] = np.arctan2(data.T_w_cam0[j][1,0], data.T_w_cam0[j][0,0])
        
        #Array with total distances traveled and changes in angles
        for k in range(len(data.T_w_cam0)-1):
            Y_delta[k+1,0] = Y[k+1,0]
            Y_delta[k+1,1] = Y[k+1,1]
            Y_delta[k+1,2] = Y[k+1,2]
            Y_delta[k+1,3] = Y[k+1,3]-Y[k,3] #The angles varied from -pi to pi, so it seemed global
            Y_delta[k+1,4] = Y[k+1,4]-Y[k,4] #This is calculating differences
            Y_delta[k+1,5] = Y[k+1,5]-Y[k,5]
        
        #Array with delta distances, delta angles, and euclidian distance
        for k in range(len(data.T_w_cam0)-1):
            Y_d[k+1,0] = Y[k+1,0]-Y[k,0]
            Y_d[k+1,1] = Y[k+1,1]-Y[k,1]
            Y_d[k+1,2] = Y[k+1,2]-Y[k,2]
            Y_d[k+1,3] = Y[k+1,3]-Y[k,3] 
            Y_d[k+1,4] = Y[k+1,4]-Y[k,4] 
            Y_d[k+1,5] = Y[k+1,5]-Y[k,5]
            Y_d[k+1,6] = np.sqrt(np.square(Y_d[k+1,0])+np.square(Y_d[k+1,1])+np.square(Y_d[k+1,2]))
        
        list_Y.append(Y)
        
    return list_Y

d = load_trans_rot_matrix()
q = d[2]

#t = np.arange(q.shape[0])
#s = np.arange(100)
#plt.plot(t,q[:,2],t,q[:,6]*np.cos(q[:,5]))
#plt.plot(q[:,0],q[:,2])
#plt.show()
```
