```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
model_dir = '/home/sexy/source/deep-visual-odometry/models/'
netNum = '7'

with open(model_dir+"alexNet_"+netNum+'/eval_S.json') as data_file:    
    data_S = json.load(data_file)
with open(model_dir+"alexNet_"+netNum+'/eval_Tr.json') as data_file:    
    data_Tr = json.load(data_file)
with open(model_dir+"alexNet_"+netNum+'/eval_Te.json') as data_file:    
    data_Te = json.load(data_file)

plot_data = [[],[],[]]
for i in range(11):
    plot_data[0].append(np.asarray(data_S[i]))
    plot_data[1].append(np.asarray(data_Tr[i]))
    plot_data[2].append(np.asarray(data_Te[i]))
```

```python
# Blue is the ground truth
# Orange is the predicted data

# This plots the full sequence (Training data with Test data appended)
plt.figure(1, figsize=(16,12))
for i in range(11):
    plt.subplot(3,4,i+1)
    plt.plot(plot_data[0][i][:,0], plot_data[0][i][:,1], plot_data[0][i][:,2], plot_data[0][i][:,3])
plt.show()
```

```python
# Training data
plt.figure(2, figsize=(16,12))
for i in range(11):
    plt.subplot(3,4,i+1)
    plt.plot(plot_data[1][i][:,0], plot_data[1][i][:,1], plot_data[1][i][:,2], plot_data[1][i][:,3])
plt.show()
```

```python
# Test data
plt.figure(3, figsize=(16,12))
for i in range(11):
    plt.subplot(3,4,i+1)
    plt.axis('equal')
    plt.plot(plot_data[2][i][:,0], plot_data[2][i][:,1], plot_data[2][i][:,2], plot_data[2][i][:,3])
plt.show()
```

```python
# This plots the full sequence (Training data with Test data appended)
plt.figure(4, figsize=(16,12))
for i in range(11):
    plt.subplot(3,4,i+1)
    plt.axis('equal')
    plt.plot(plot_data[0][i][:,0], plot_data[0][i][:,1], plot_data[0][i][:,2], plot_data[0][i][:,3])
plt.show()
```
